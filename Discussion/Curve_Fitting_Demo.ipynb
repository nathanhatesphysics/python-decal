```python
import matplotlib.pyplot as plt
import scipy.optimize as fit
import numpy as np
```

```python
m = 0.5
b = 7
sig = 1
n = 20
xdata = np.linspace(0, 10, 20)
ydata = (m * xdata + b) + np.random.standard_normal(size=n) * sig
```

```python
plt.figure()
plt.scatter(xdata, ydata)
plt.show()
```

```python
def f(x, m, b):
    return m * x + b

p0 = [1, 5]

parameters, covariance_matrix = fit.curve_fit(f, xdata, ydata, p0)
```
