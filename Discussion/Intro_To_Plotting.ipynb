```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure()
plt.plot(x, y, color='mediumspringgreen', marker='1', linewidth=30)
plt.scatter(x, y2, color='mediumturquoise', marker='8', linewidth=40)
plt.plot(x, y2*2, color='lightpink', linestyle='dashdot', linewidth=5)
plt.title("sinx vs cosx")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["sin", "cos", "cos*2"])
plt.show
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[1, 1].plot(x, y2)
```
