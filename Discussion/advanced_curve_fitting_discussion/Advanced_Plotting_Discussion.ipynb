### Question 1: Pillars of Creation!

In this question, we're going to visualize real world data using ImShow. Specifically, we'll be plotting data that represents the Pillars of Creation (Eagle Nebula/Messier 16). Things that you will have to do...

### Question 3: Scaling Plots

In this question, we're going to see how changing the scale on a plot will affect the way we view and analyze data.
To do this, we're going to use some generated data.
- Create a figure with 2 subplots.
- On the right side, plot the data (scaling_data). Do you notice something about the y-axis scale?
- On the left side, plot this same data with a readjusted y-axis scale. (you can use a matplotlib operation, or manually rescale the data, and then plot it)

```python
#Data
scaling_data = [10**(0.1*i) for i in range(100)]
```

```python
#Create your plots here!
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,4))
ax1.plot(scaling_data)
ax1.set_title("normal")
ax2.plot(scaling_data)
ax2.set_title('log scale')
ax2.set_yscale('log')
plt.show()
```
