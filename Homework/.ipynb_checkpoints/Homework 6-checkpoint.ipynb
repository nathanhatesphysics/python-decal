```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1 Create function
def f(x, A, w):
    return A * np.sin(w * x)
```

```python
x_data = np.linspace(0, 2*np.pi, 1000)
```

```python
A_data = [1, 2, 3, 4, 5]
w_data = [1, 2, 3, 4, 5]
```

```python
plt.figure()
for i in range(0, len(A_data)):
    plt.plot(x_data, f(x_data, A_data[i], w_data[i]))
plt.title("sinusoidal functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend([1, 2, 3, 4, 5])
plt.show()
```
