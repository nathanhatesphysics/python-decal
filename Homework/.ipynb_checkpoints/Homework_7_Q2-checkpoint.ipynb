```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
planets = sns.load_dataset('planets')
```

```python
planets.head(5)
```

```python
sns.set_theme()
```

```python
sns.relplot(
    data = planets,
    x = 'orbital_period',
    y = 'mass',
    hue = 'method', style = 'method', size = 2
)
plt.yscale('log')
plt.xscale('log')
plt.title('orbital period vs. mass')
plt.show()
# i don't know what the units are, but if i did i would add them to the axis titles
```

```python
new_planets = planets.dropna(subset=['year'])
```

```python
sns.histplot(
    data = new_planets,
    x = 'year',
    hue = 'method'
)
```
